```python
import re

new_name_data = []
new_name_data_original = []
new_abbreviation_scores = dict()
new_common_abbreviations = []

# Function created to load the files
def load_abbreviation_scores():
    file = open("sharma_values.txt", "r")
    while True:
        line = file.readline()
        if len(line) == 0:
            break
        else:
            temp = list(map(str, line.split()))
            new_abbreviation_scores[temp[0]] = int(temp[1])
    file.close()

# Assigning the different scores
def calculate_abbreviation_score(abbrev, name_list):
    first, last, score = "", "", 0
    for i in name_list:
        if len(i) > 1:
            last += i[-1]
        first += i[0]

    for j in abbrev[1:]:
        if j in first:
            score += 0
            first = first[first.index(j):]

        elif j in last:
            if j == "E":
                score += 20
            else:
                score += 5
            last = last[last.index(j):]
        else:
            l = []
            for i in name_list:
                if j in i:
                    l.append(min(i.index(j), 3))
            score += min(l) + new_abbreviation_scores[j]
    return score

# Creating the abbreviations from the names
def generate_abbreviations(name_list):
    name = "".join(name_list)
    n = len(name)
    abbrev = dict()
    for i in range(1, n):
        for j in range(i + 1, n):
            temp = name[0] + name[i] + name[j]
            if temp in new_common_abbreviations:
                continue
            else:
                abbrev[temp.upper()] = calculate_abbreviation_score(temp, name_list)

    return abbrev

if __name__ == "__main__":
    new_name = input("Enter the file name: ")

    new_file = open(new_name, "r")
    load_abbreviation_scores()

    while True:
        line = new_file.readline()
        if len(line) == 0:
            break
        else:
            new_name_data.append(list(map(str, re.split(" |,|-|'", line.upper().strip()))))
            new_name_data_original.append(list(map(str, re.split(" |,|-|'", line.strip()))))

    new_list = []

    for i in new_name_data:
        new_list.append(set(generate_abbreviations(i).keys()))

    # Saving our final result in a different file named as New_Result.txt
    for i in range(len(new_list)):
        for j in range(i + 1, len(new_list)):
            temp = new_list[i].intersection(new_list[j])
            if len(temp) > 0:
                new_common_abbreviations += temp
    new_out = open("sharma_trees_abbrevs.txt", "w")
    
    # This will store the name with the smallest calculated score abbreviation
    for i in range(len(new_name_data)):
        temp = generate_abbreviations(new_name_data[i])
        new_out.write(" ".join(new_name_data_original[i]) + "\n")

        if len(temp) > 0:
            m = min(temp.items(), key=lambda x: x[1])[1]
            for j in temp.keys():
                if temp[j] == m:
                    new_out.write(j + " ")
        new_out.write("\n")
    new_out.close()
    new_file.close()
    print("Task completed successfully: Output file is save as 'sharma_trees_abbrevs.txt'")

# Opening the file New_Result.txt
new_data_file = open('sharma_trees_abbrevs.txt', 'r')
new_data = new_data_file.read()
new_result = new_data.split('\n')

# Creating two lists
new_alpha = []
new_digits = []
i = 0
while i < len(new_result):
    if i % 2 == 0:
        new_alpha.append(new_result[i])
    else:
        new_digits.append(new_result[i])
    i = i + 1

# Converted the two lists into a list of tuples
new_list_of_tuples = list(zip(new_alpha, new_digits))
# Converted the list of tuples into a dictionary
new_dict = dict(new_list_of_tuples)
print(new_dict)

# for k, v in new_dict.items():
#     print(k, " : ", v)
```
